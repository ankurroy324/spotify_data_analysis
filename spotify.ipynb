```python
import pandas as pd
```

```python
df = pd.read_csv('music.csv')
print(df.head(20))
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df["artist_name"] = df["artist_name"].transform(lambda x: x.fillna("unknown_artist"))
```

```python
df["artist_name"].isnull().sum()
```

```python
df["artist_genres"] = df["artist_genres"].transform(lambda x : x.fillna("unclassified"))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df['album_release_date'] = pd.to_datetime(df['album_release_date'],format = 'mixed',errors = 'coerce')

df['release_year'] = df['album_release_date'].dt.year
df['release_month'] = df['album_release_date'].dt.month
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['primary_genre'] = df['artist_genres'].astype(str).str.strip('[]').str.split(',').str[0]
```

```python
df['primary_genre']
```

```python
def categorize_popularity(score):
    if score >= 80:
        return 'Popular'
    elif score >= 50:
        return 'Moderate'
    else:
        return 'Low'

   
```

```python
 df['popularity_category'] = df['track_popularity'].apply(categorize_popularity)
```
